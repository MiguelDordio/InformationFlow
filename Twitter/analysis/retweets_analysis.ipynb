```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import plotly.express as px
```

```python
sns.set()
```

```python
DATASET_PATH = "../../data/tweets_2020_2021_v2.csv"
RETWEETS_PATH = "../../data/retweets.csv"
RETWEETS_USERS_PATH = "../../data/retweets_users.csv"
DATASET_ANALYSIS_2020_PATH = "../../data/tweets_analysis_2020.csv"
DATASET_ANALYSIS_2021_PATH = "../../data/tweets_analysis_2021.csv"
```

```python
df = pd.read_csv(filepath_or_buffer=DATASET_PATH, sep=",")
df = df.sort_values(by='timestamp', ascending=True)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.drop(['index', 'Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.2'], axis=1)
df.tail()
```

```python
df_retweets = pd.read_csv(filepath_or_buffer=RETWEETS_PATH, sep=",")
df_retweets['timestamp'] = pd.to_datetime(df_retweets['timestamp'])
df_retweets = df_retweets.sort_values(by='timestamp', ascending=True).reset_index()
df_retweets
```

```python
df_retweeters = pd.read_csv(filepath_or_buffer=RETWEETS_USERS_PATH, sep=",")
df_retweeters['created_at'] = pd.to_datetime(df_retweeters['created_at'])
df_retweeters = df_retweeters.drop_duplicates(subset=['user_id'], keep="first")
df_retweeters
```

```python
df_retweets_info = pd.merge(df_retweets, df_retweeters[['user_id', 'followers', 'following', 'tweet_count', 'verified', 'created_at']], how='left', on="user_id")
df_retweets_info = df_retweets_info.drop(['index', 'Unnamed: 0'], axis=1)
df_retweets_info.head(5)
```

### Removing miss colected retweets

```python
df_retweets_info[pd.isna(df_retweets_info['referenced_tweets'])]
```

```python
df_retweets_info = df_retweets_info[df_retweets_info['referenced_tweets'].notna()]
```

### Extracting tweet ID from referenced_tweets field

```python
def get_ref_tweet_id(ref_tweets):
    g = str(ref_tweets)
    if isinstance(g, str) and g != 'nan' and not pd.isna(g):
        return int(g.split()[1].replace(',', ''))
    else:
        return None
```

```python
df_retweets_info['ref_tweed_id'] = df_retweets_info['referenced_tweets'].apply(lambda ref_tweet: get_ref_tweet_id(ref_tweet))
df_retweets_info.head()
```

## Analysing average followers count between shared tweets and not shared by topic

```python
topics_categories = df['topics_cleaned'].unique()[1:]
cats_vals = []
for category in topics_categories:
    tweets_by_topic = df[df['topics_cleaned'] == category]
    shared_tweets_avg = tweets_by_topic[tweets_by_topic['retweet_count'] > 0]['followers'].mean()
    non_shared_tweets_avg = tweets_by_topic[tweets_by_topic['retweet_count'] == 0]['followers'].mean()
    cat_vals = dict()
    cat_vals['Type'] = 'Shared'
    cat_vals['Topic'] = category
    cat_vals['Average'] = int(shared_tweets_avg)
    cats_vals.append(cat_vals)

    cat_vals = dict()
    cat_vals['Type'] = 'Not shared'
    cat_vals['Topic'] = category
    cat_vals['Average'] = int(non_shared_tweets_avg)
    cats_vals.append(cat_vals)

print(cats_vals)
df_test1 = pd.DataFrame(cats_vals, columns=['Type', 'Topic', 'Average'])
df_test1
```

```python
fig = px.bar(df_test1, x="Topic", y="Average", color="Type", color_discrete_sequence=px.colors.qualitative.Dark24, barmode="group",
             title="Analysing average followers count between shared tweets and not shared by topic")
fig.show()
```

## Analysing average followers count of retweeters per topic

```python
tweets_by_topic = df[(df['topics_cleaned'] == 'Person') & (df['retweet_count'] > 0)]
tweets_by_topic
```

```python
def get_avg_followers_retweets(og_tweet_id):
    matches = df_retweets_info[df_retweets_info['ref_tweed_id'] == og_tweet_id]
    if matches.shape[0] == 0: return -1
    return int(matches['followers'].mean())
```

```python
tweets_by_topic['avg_retweeters_follower'] = tweets_by_topic['tweet_id'].apply(lambda x: get_avg_followers_retweets(x))
tweets_by_topic
```

```python
tweets_by_topic[tweets_by_topic['avg_retweeters_follower'] == -1]
```

## Analysing average retweet time ratio of retweeters per topic

```python
df_retweets_info['timestamp'] = pd.to_datetime(df_retweets_info['timestamp'])
```

```python
def get_avg_retweets_time(og_tweet_time, og_tweet_id):
    matches = df_retweets_info[df_retweets_info['ref_tweed_id'] == og_tweet_id]
    if matches.shape[0] == 0: return -1
    avg_time_diff = matches['timestamp'].apply(lambda retweet_time: abs(og_tweet_time.replace(tzinfo=datetime.timezone.utc) - retweet_time.replace(tzinfo=datetime.timezone.utc)).total_seconds() / 3600.0).mean()
    return int(avg_time_diff)
```

```python
tweets_by_topic['avg_retweeters_time'] = [get_avg_retweets_time(x, y) for x, y in zip(tweets_by_topic['timestamp'], tweets_by_topic['tweet_id'])]
tweets_by_topic
```

## Analysing average retweeter's account age per topic

```python
df_retweets_info['created_at'] = pd.to_datetime(df_retweets_info['created_at'])
```

```python
def get_avg_retweets_account_age(og_tweet_id):
    matches = df_retweets_info[df_retweets_info['ref_tweed_id'] == og_tweet_id]
    if matches.shape[0] == 0: return -1
    avg_time_diff = matches['timestamp'].apply(lambda created_at: datetime.datetime.now().replace(tzinfo=datetime.timezone.utc) - created_at.replace(tzinfo=datetime.timezone.utc)).mean()
    return round((avg_time_diff / np.timedelta64(1, 'Y')), 3)
```

```python
tweets_by_topic['avg_retweeters_account_age'] = [get_avg_retweets_account_age(x) for x in tweets_by_topic['tweet_id']]
tweets_by_topic
```

```python
df_test3 = tweets_by_topic[tweets_by_topic['avg_retweeters_account_age'] != -1]
df_test3['avg_retweeters_account_age'] = [round(x, 3) for x in df_test3['avg_retweeters_account_age']]
```

```python
fig = px.bar(df_test3, x="topics_cleaned", y="avg_retweeters_account_age", color="topics_cleaned", color_discrete_sequence=px.colors.qualitative.Dark24,
             title="Analysing average followers count between shared tweets and not shared by topic")
fig.show()
```

```python
cats_vals = []
for category in topics_categories:
    tweets_by_topic = df[(df['topics_cleaned'] == category) & (df['retweet_count'] > 0)]

    tweets_by_topic['avg_retweeters_followers'] = [get_avg_followers_retweets(x) for x in zip(tweets_by_topic['tweet_id'])]
    tweets_by_topic['avg_retweeters_time'] = [get_avg_retweets_time(x, y) for x, y in zip(tweets_by_topic['timestamp'], tweets_by_topic['tweet_id'])]
    tweets_by_topic['avg_retweeters_account_age'] = [get_avg_retweets_account_age(x) for x in tweets_by_topic['tweet_id']]

    avg_retweeters_followers = tweets_by_topic['avg_retweeters_followers'].mean()
    avg_retweeters_time = tweets_by_topic['avg_retweeters_time'].mean()
    avg_retweeters_account_age = tweets_by_topic['avg_retweeters_account_age'].mean()

    cat_vals = dict()
    cat_vals['Topic'] = category
    cat_vals['Average Retweeters Followers'] = int(avg_retweeters_followers)
    cat_vals['Average Retweets Time'] = int(avg_retweeters_time)
    cat_vals['Average Retweeters Account Age'] = int(avg_retweeters_account_age)
    cats_vals.append(cat_vals)
```

```python
print(cats_vals)
df_test3 = pd.DataFrame(cats_vals, columns=['Topic', 'Average Retweeters Followers', 'Average Retweets Time', 'Average Retweeters Account Age'])
df_test3
```

```python
fig = px.bar(df_test3, x="Topic", y="Average", color="Type", color_discrete_sequence=px.colors.qualitative.Dark24, barmode="group",
             title="Analysing average followers count between shared tweets and not shared by topic")
fig.show()
```
